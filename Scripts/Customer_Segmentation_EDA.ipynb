# Exploratory Data Analysis (EDA)

```python
# Import necessary libraries for EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Ecommerce_Consumer_Behavior_cleaned_dataset.csv', parse_dates=['Time_of_Purchase'])
```

```python
# Set style for plots
sns.set(style="whitegrid")
```

```python
# Identify columns by type
numeric_cols = [
    'Age', 'Purchase_Amount', 'Frequency_of_Purchase', 
    'Brand_Loyalty', 'Product_Rating', 'Return_Rate', 
    'Customer_Satisfaction', 'Time_Spent_on_Product_Research(hours)', 
    'Time_to_Decision'
]

categorical_cols = [
    'Gender', 'Income_Level', 'Marital_Status', 'Education_Level', 
    'Occupation', 'Purchase_Category', 'Purchase_Channel', 
    'Social_Media_Influence', 'Discount_Sensitivity', 
    'Engagement_with_Ads', 'Device_Used_for_Shopping', 
    'Payment_Method', 'Purchase_Intent', 'Shipping_Preference'
]

bool_cols = ['Discount_Used', 'Customer_Loyalty_Program_Member']

datetime_cols = ['Time_of_Purchase']
```

## Univariate Analysis

### Demographic Profile

```python
#number of customers by age range
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Age', bins=5, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
## Demographics: Categorical variables
demographic_cols = [
    'Gender', 'Income_Level', 'Marital_Status', 'Education_Level', 
    'Occupation'
]

for col in demographic_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette='pastel', hue=col, legend=False)
    plt.title(f'{col}')
    plt.xlabel('')
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=90)
    plt.show()
```

### Multivariate Analysis

##### Do older clients spend more, and how does this relate to their income level?

```python
#Age vs. Purchase Amount + Income Level
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df, 
    x='Age', 
    y='Purchase_Amount', 
    hue='Income_Level', 
    size='Frequency_of_Purchase',  # Third dimension
    palette='viridis',
    alpha=0.7
)
plt.title("Age vs. Purchase Amount (Color=Income, Size=Purchase Frequency)")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

##### What product categories do different income levels prefer?

```python
#Income Level vs. Purchase Category
plt.figure(figsize=(14, 7))
sns.boxplot(
    data=df,
    x='Purchase_Category',
    y='Purchase_Amount',
    hue='Income_Level',
    palette='Set2'
)
plt.xticks(rotation=90)
plt.title("Purchase Category Preferences by Income Level")
plt.xlabel("Product Category")
plt.ylabel("Purchase Amount (USD)")
plt.legend(title='Income Level')
plt.show()
```

##### Are there gender differences in shopping channel preference?

```python
# Gender vs. Purchase Channel
plt.figure(figsize=(12, 6))
sns.countplot(
    data=df,
    x='Purchase_Channel',
    hue='Gender',
    palette='pastel',
    edgecolor='black'
)
plt.title("Purchase Channel Preference by Gender")
plt.xlabel("Purchase Channel")
plt.ylabel("Count")
plt.show()
```

##### Are there gender differences in sensitivity to discounts?

```python
# Gender vs Discount Sensitivity
# Discount Sensitivity by Gender
import numpy as np
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df,
    x='Discount_Sensitivity',
    y='Purchase_Amount',
    hue='Gender',
    estimator=np.mean,
    errorbar=None
)
plt.title("Average Spending by Discount Sensitivity and Gender")
plt.show()
```

##### Do brand loyal customers report higher satisfaction?

```python
# Brand Loyalty vs. Customer Satisfaction
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=df,
    x='Brand_Loyalty',
    y='Customer_Satisfaction',
    hue='Brand_Loyalty',
    palette='Set2'
)
plt.xticks()
plt.title("Brand Loyalty vs. Customer Satisfaction")
plt.xlabel("Brand_Loyalty")
plt.ylabel("Customer_Satisfaction")
plt.legend([])
plt.show()
```

##### Does the influence of social networks affect the type of purchase (impulsive vs. planned)?

```python
plt.figure(figsize=(12, 6))
sns.heatmap(
    pd.crosstab(df['Social_Media_Influence'], df['Purchase_Intent'], normalize='index'),
    annot=True,
    cmap='Blues',
    fmt='.1%'
)
plt.title("Purchase Intent Distribution by Social Media Influence")
plt.xlabel("Purchase Intent")
plt.ylabel("Social Media Influence")
plt.show()
```

##### Are payment methods associated with purchase frequency?

```python
#Payment Method vs. Purchase Frequency
plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df,
    x='Payment_Method',
    y='Frequency_of_Purchase',
    palette='muted',
    hue='Payment_Method',
    inner='quartile'
)
plt.title("Purchase Frequency Distribution by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

##### Do loyalty program members use fewer discounts?

```python
# Discount Used vs. Loyalty Program Membership
plt.figure(figsize=(8, 5))
sns.heatmap(
    pd.crosstab(df['Discount_Used'], df['Customer_Loyalty_Program_Member']),
    annot=True,
    cmap='Greens',
    fmt='d'
)
plt.title("Discount Usage vs. Loyalty Program Membership")
plt.xlabel("Loyalty Program Member")
plt.ylabel("Discount Used")
plt.show()
```

##### Do the devices used vary by purchase channel?

```python
# Purchase Channel vs. Device Used
plt.figure(figsize=(10, 6))
ax = sns.countplot(
    data=df,
    x='Purchase_Channel',
    hue='Device_Used_for_Shopping',
    palette='rocket'
)

plt.title("Device Usage by Purchase Channel")
plt.xlabel("Purchase Channel")
plt.ylabel("Count")

# Mover la leyenda fuera del gráfico a la derecha
plt.legend(
    title='Device Used',          # Título de la leyenda
    bbox_to_anchor=(1.05, 1),     # Posición relativa (x, y) fuera del área del gráfico
    loc='upper left',             # Punto de anclaje de la leyenda
    borderaxespad=0.1             # Espacio entre el gráfico y la leyenda
)

plt.tight_layout()  # Ajusta el layout para evitar cortes
plt.show()
```

### Basic distribution over time

```python
import calendar

plt.figure(figsize=(10, 5))

# Ensure that we have every month (1-12) even with 0 purchases
monthly_counts = df['Time_of_Purchase'].dt.month.value_counts()\
                                        .reindex(range(1, 13), fill_value=0)

ax = monthly_counts.plot(kind='bar', color='slateblue')

# Average line
promedio = monthly_counts.mean()
ax.axhline(promedio, color='red', linestyle='--', linewidth=1.5, label='Monthly Average')

# Set up month names
meses = list(calendar.month_abbr)[1:]  # ['Jan', 'Feb', ..., 'Dec']
ax.set_xticks(range(12))  # Positions 0-11 for 12 months
ax.set_xticklabels(meses, rotation=45, ha='right') 

plt.title('Number of Purchases by Month')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()  # Automatic layout adjustment
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Calculate monthly sum of ‘Purchase_Amount’ and insure all months (1-12)
monthly_sums = (
    df.groupby(df['Time_of_Purchase'].dt.month)['Purchase_Amount']
    .sum()
    .reindex(range(1, 13), fill_value=0)  # Include months with no sales
)

ax = monthly_sums.plot(kind='bar', color='slateblue')

# Average line
promedio = monthly_sums.mean()
ax.axhline(promedio, color='red', linestyle='--', linewidth=1.5, label='Monthly Average')

# Set up month names
meses = list(calendar.month_abbr)[1:]  
ax.set_xticks(range(12)) 
ax.set_xticklabels(meses, rotation=45, ha='right')

plt.title('Total Purchase Amount by Month')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount ($)')  # Adjust currency unit if necessary
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
